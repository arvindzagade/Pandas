# Rank Values with .rank() Method

```python
# Removing all null values

nba = pd.read_csv(r'.\pandas\nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.head(3)
```

```python
## Rank is called on single series...suppose giving a player with higest salary as rank of 1 and so on..

nba['Salary']
```

```python
nba['Salary'].rank() ## this is ranking it in revers as 0 is getting a 6th position
                    ## so highest salary is ranked toward the end
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')
```

```python
nba.head(3)
```

```python
#Checking if we are getting rank according to salary

nba.sort_values(by='Salary').head()
```

```python
nba.sort_values(by='Salary',ascending=False).head()
```
