# Create A Random Sample from a Dataframe

```python
bond = pd.read_csv('.\pandas\jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.sample()
```

```python
bond.sample(n=5)
```

```python
# if we want to get fraction(eg.25%) Data from Dataframe.will use frac parameter

bond.sample(frac=.25)
```

```python
## All these methods are same for getting rows
bond.sample(n=3,axis=0)
bond.sample(n=3,axis='index')
```

```python
## All these methods are same for getting columns
bond.sample(n=3,axis=1)
bond.sample(n=3,axis='columns')
```

# .where Method()

```python
bond = pd.read_csv('.\pandas\jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
mask = bond['Actor'] == 'Sean Connery'
bond[mask]
```

```python
bond.where(mask)
```

```python
bond.where(bond['Box Office'] > 800)
```

```python
mask2 = bond['Box Office'] > 800
```

```python
bond.where(mask & mask2)
```

# Query Method

```python
bond = pd.read_csv('.\pandas\jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.columns
```

```python
[column_name.replace(' ','_') for column_name in bond.columns]
```

```python
bond.columns = [column_name.replace(' ','_') for column_name in bond.columns]
bond.head(1)
```

```python
bond.query('Actor == "Sean Connery"')
```

```python
bond.query('Director == "Terence Young"')
```

```python
bond.query('Actor != "Roger Moore"')
```

```python
bond.query("Box_Office > 600")
```

```python
bond.query('Actor == "Roger Moore" and Director=="John Glen"')
```

```python
bond.query('Actor in["Timothy Dalton","Roger Moore"]')
```

```python
bond.query('Actor not in["Timothy Dalton","Roger Moore"]')
```

# .apply() Method on single column

```python
bond = pd.read_csv('.\pandas\jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_to_string_and_add_milliion(number):
    return str(number) + ' Millions'
bond['Box Office'].apply(convert_to_string_and_add_milliion)
```
